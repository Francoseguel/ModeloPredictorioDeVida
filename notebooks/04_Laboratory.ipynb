## BIOPRO_J (Standard Biochemistry)

```python
df = catalog.load("biopro_j_dataset")
df
```

```python
print(f"Shape: {df.shape}")
df.head()
```

```python
df.tail()
```

```python
df.dtypes.to_frame("dtype")
```

```python
import pandas as pd
nulls = df.isnull().sum()
null_pct = (nulls / len(df) * 100).round(2)
null_df = pd.DataFrame({"nulos": nulls, "porcentaje (%)": null_pct})
null_df.sort_values("porcentaje (%)", ascending=False)
```

```python
import matplotlib.pyplot as plt
import math

num_cols = df.select_dtypes(include="number").columns.tolist()
n = len(num_cols)
cols_per_row = 4
rows = math.ceil(n / cols_per_row)
fig, axes = plt.subplots(rows, cols_per_row, figsize=(cols_per_row*4, rows*3))
axes = axes.flatten() if n > 1 else [axes]
for i, col in enumerate(num_cols):
    df[col].dropna().plot(kind="hist", bins=30, ax=axes[i], title=col, color="steelblue", edgecolor="white")
    axes[i].set_xlabel("")
for j in range(len(num_cols), len(axes)):
    axes[j].set_visible(False)
plt.suptitle("Histogramas — variables numéricas", fontsize=14, y=1.01)
plt.tight_layout()
plt.show()
```

```python
num_cols = df.select_dtypes(include="number").columns.tolist()
n = len(num_cols)
cols_per_row = 4
rows = math.ceil(n / cols_per_row)
fig, axes = plt.subplots(rows, cols_per_row, figsize=(cols_per_row*4, rows*3))
axes = axes.flatten() if n > 1 else [axes]
for i, col in enumerate(num_cols):
    df.boxplot(column=col, ax=axes[i])
    axes[i].set_title(col)
for j in range(len(num_cols), len(axes)):
    axes[j].set_visible(False)
plt.suptitle("Boxplots — detección de outliers", fontsize=14, y=1.01)
plt.tight_layout()
plt.show()
```
