# Linked Mortality File (LMF) — Carga

Archivo: `NHANES_2017_2018_MORT_2019_PUBLIC.dat`  
Seguimiento hasta: **31 diciembre 2019**  
Fuente: NCHS / CDC · Publicado mayo 2022

El `.dat` es ancho fijo (fixed-width). Los `colspecs` vienen del `SAS_ReadInProgramAllSurveys.sas` (sección NHANES).

```python
import pandas as pd
from pathlib import Path

DAT_FILE = Path("../data/01_raw/NHANES_2017_2018_MORT_2019_PUBLIC.dat")
```

```python
# El .dat es ancho fijo: cada variable ocupa posiciones fijas de carácter.
# colspecs indica (inicio, fin) en base-0 (el SAS usa base-1, por eso se resta 1 al inicio).
# Fuente: sección NHANES del SAS_ReadInProgramAllSurveys.sas
colspecs = [
    (0,  6),   # SEQN          cols 1-6   — ID del participante, llave de join con todos los .xpt
    (14, 15),  # ELIGSTAT      col 15     — 1=Eligible, 2=Menor de 18, 3=Inelegible
    (15, 16),  # MORTSTAT      col 16     — 1=Fallecido, 0=Vivo/censurado al 31-dic-2019
    (16, 19),  # UCOD_LEADING  cols 17-19 — Causa principal de muerte ('001'=corazón, '002'=cáncer, etc.)
    (19, 20),  # DIABETES      col 20     — 1=Diabetes mencionada en certificado de defunción
    (20, 21),  # HYPERTEN      col 21     — 1=Hipertensión mencionada en certificado de defunción
    (42, 45),  # PERMTH_INT    cols 43-45 — Meses de seguimiento desde la entrevista hasta muerte/censura
    (45, 48),  # PERMTH_EXM    cols 46-48 — Meses de seguimiento desde el examen físico (MEC) hasta muerte/censura
]

names = [
    "SEQN",
    "ELIGSTAT",
    "MORTSTAT",
    "UCOD_LEADING",
    "DIABETES",
    "HYPERTEN",
    "PERMTH_INT",
    "PERMTH_EXM",
]

# UCOD_LEADING se lee como str para preservar los ceros a la izquierda ('001', '002', etc.)
lmf = pd.read_fwf(
    DAT_FILE,
    colspecs=colspecs,
    names=names,
    dtype={"SEQN": int, "UCOD_LEADING": str},
)

print(f"Filas: {len(lmf):,}")
lmf.head()
```
